```python
import os
import shutil
import glob
import random
from tqdm import tqdm
from google.colab import drive

if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

real_base_path = '/content/drive/MyDrive/PBL/OCTID'


syn_base_path = '/content/drive/MyDrive/Syn_Dataset'

output_base = '/content/drive/MyDrive/Final_Dataset_For_CNN'


classes = ['CSR', 'DR', 'MH', 'NORMAL']
split_ratio = 0.2

if not os.path.exists(syn_base_path):
    print(f"âŒ ERROR: Could not find '{syn_base_path}'")
    print("Please check your Drive. Did you name it 'Syn_Dataset' or just 'Syn'?")

    raise SystemExit("Stopping script to prevent errors.")

def create_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)


if os.path.exists(output_base):
    print(f"ğŸ§¹ Removing old dataset at {output_base}...")
    shutil.rmtree(output_base)

print(f"ğŸš€ Building Final Dataset using GOOD data from: {syn_base_path}")

for category in classes:
    print(f"\nProcessing: {category}")


    real_source = os.path.join(real_base_path, category)
    syn_source = os.path.join(syn_base_path, category)


    train_dest = os.path.join(output_base, 'train', category)
    test_dest = os.path.join(output_base, 'test', category)
    create_dir(train_dest)
    create_dir(test_dest)


    real_imgs = []
    for ext in ['*.png', '*.jpg', '*.jpeg', '*.PNG', '*.JPG']:
        real_imgs.extend(glob.glob(os.path.join(real_source, ext)))

    random.shuffle(real_imgs)

    if len(real_imgs) == 0:
        print(f"   âŒ WARNING: No Real images found for {category}!")
    else:
        split = int(len(real_imgs) * split_ratio)
        test_files = real_imgs[:split]
        train_files = real_imgs[split:]

        for f in test_files: shutil.copy(f, test_dest)
        for f in train_files: shutil.copy(f, train_dest)
        print(f"   - Real Images: {len(test_files)} sent to Test, {len(train_files)} sent to Train")


    syn_imgs = []
    for ext in ['*.png', '*.jpg', '*.jpeg', '*.PNG', '*.JPG']:
        syn_imgs.extend(glob.glob(os.path.join(syn_source, ext)))

    if len(syn_imgs) > 0:
        for f in tqdm(syn_imgs, desc="   - Copying Synthetic"):
            shutil.copy(f, train_dest)
    else:
        print(f"   âš ï¸ WARNING: No Synthetic images found for {category} in {syn_source}")

print("\nâœ… SUCCESS! Dataset is ready for the CNN.")
```

```python
import os
import shutil
import glob
import random
from tqdm import tqdm


target_class = 'NORMAL'

real_base_path = '/content/drive/MyDrive/PBL/OCTID'
syn_base_path = '/content/drive/MyDrive/Syn_Dataset'
output_base = '/content/drive/MyDrive/Final_Dataset_For_CNN'


print(f"ğŸš‘ STARTING RESCUE MISSION FOR: {target_class}")


real_source = os.path.join(real_base_path, target_class)
syn_source = os.path.join(syn_base_path, target_class)


train_dest = os.path.join(output_base, 'train', target_class)
test_dest = os.path.join(output_base, 'test', target_class)


if not os.path.exists(train_dest):
    os.makedirs(train_dest)
    print(f"   âœ… Created missing folder: {train_dest}")
if not os.path.exists(test_dest):
    os.makedirs(test_dest)
    print(f"   âœ… Created missing folder: {test_dest}")


real_imgs = []

for ext in ['*.png', '*.jpg', '*.jpeg', '*.PNG', '*.JPG']:
    real_imgs.extend(glob.glob(os.path.join(real_source, ext)))

print(f"\nğŸ” Found {len(real_imgs)} REAL images in source.")

if len(real_imgs) == 0:
    print(f"âŒ ERROR: Could not find real images at: {real_source}")
    print("ğŸ‘‰ Check: Is the folder named 'Normal' instead of 'NORMAL'?")
else:
    random.shuffle(real_imgs)
    split = int(len(real_imgs) * 0.2)

    test_files = real_imgs[:split]
    train_files = real_imgs[split:]

    print(f"   - Moving {len(test_files)} to TEST...")
    for f in test_files: shutil.copy(f, test_dest)

    print(f"   - Moving {len(train_files)} to TRAIN...")
    for f in train_files: shutil.copy(f, train_dest)


syn_imgs = []
for ext in ['*.png', '*.jpg', '*.jpeg', '*.PNG', '*.JPG']:
    syn_imgs.extend(glob.glob(os.path.join(syn_source, ext)))

print(f"\nğŸ” Found {len(syn_imgs)} SYNTHETIC images in source.")

if len(syn_imgs) == 0:
    print(f"âŒ ERROR: Could not find synthetic images at: {syn_source}")
else:
    print(f"   - Moving {len(syn_imgs)} to TRAIN...")
    for f in tqdm(syn_imgs):
        shutil.copy(f, train_dest)

print(f"\nâœ… RESCUE COMPLETE. Check '{output_base}' again.")
```

```python
import os
import hashlib
import glob
from collections import defaultdict
from tqdm import tqdm

dataset_path = '/content/drive/MyDrive/Final_Dataset_For_CNN'
splits = ['train', 'test']
classes = ['CSR', 'DR', 'MH', 'NORMAL']

print("ğŸ¥ STARTING DATASET HEALTH CHECK...")
print(f"ğŸ“‚ Scanning: {dataset_path}\n")


file_hashes = defaultdict(list)
total_files = 0
duplicates_found = 0


def get_md5(file_path):
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()


for split in splits:
    for category in classes:
        folder_path = os.path.join(dataset_path, split, category)

        if not os.path.exists(folder_path):
            print(f"âŒ MISSING FOLDER: {folder_path}")
            continue

        files = glob.glob(os.path.join(folder_path, '*'))
        print(f"   â€¢ Scanning {split}/{category}: {len(files)} images...")

        for file_path in files:
            try:
                file_hash = get_md5(file_path)
                file_hashes[file_hash].append(file_path)
                total_files += 1
            except Exception as e:
                print(f"     âš ï¸ Error reading {file_path}")


print("\nğŸ“Š ANALYSIS RESULTS")
print("-" * 30)

leakage_count = 0
internal_dupes = 0

for file_hash, paths in file_hashes.items():
    if len(paths) > 1:
        locations = [p.split('/')[-3:] for p in paths] # ['train', 'CSR', 'img.png']


        split_types = set([loc[0] for loc in locations])
        if 'train' in split_types and 'test' in split_types:
            print(f"ğŸš¨ DATA LEAKAGE DETECTED!")
            print(f"   Hash {file_hash} found in:")
            for p in paths: print(f"    - {p}")
            leakage_count += 1
        else:

            internal_dupes += 1

print("-" * 30)
if leakage_count == 0:
    print("âœ… TEST PASSED: No Data Leakage (Train and Test are unique).")
else:
    print(f"âŒ TEST FAILED: Found {leakage_count} images shared between Train and Test.")

if internal_dupes == 0:
    print("âœ… TEST PASSED: No duplicate images found within folders.")
else:
    print(f"âš ï¸ WARNING: Found {internal_dupes} sets of duplicate images (harmless but redundant).")

print(f"\nğŸ Final Count: {total_files} unique images scanned.")
```

```python
import os
import hashlib
import glob

dataset_path = '/content/drive/MyDrive/Final_Dataset_For_CNN'
target_class = 'NORMAL'

print(f"ğŸš‘ STARTING SURGICAL REMOVAL FOR {target_class} LEAKAGE...")

train_dir = os.path.join(dataset_path, 'train', target_class)
test_dir = os.path.join(dataset_path, 'test', target_class)


test_hashes = set()
print("   Scanning Test folder (The Exam)...")
for f in glob.glob(os.path.join(test_dir, '*')):
    with open(f, "rb") as file:
        file_hash = hashlib.md5(file.read()).hexdigest()
        test_hashes.add(file_hash)

print(f"   âœ… Indexed {len(test_hashes)} unique images in Test.")

train_files = glob.glob(os.path.join(train_dir, '*'))
deleted_count = 0

print("   Scanning Train folder for duplicates...")
for f in train_files:
    with open(f, "rb") as file:
        train_hash = hashlib.md5(file.read()).hexdigest()

    if train_hash in test_hashes:

        os.remove(f)
        deleted_count += 1
        print(f"      ğŸ—‘ï¸ Deleted duplicate: {os.path.basename(f)}")

print("-" * 30)
if deleted_count > 0:
    print(f"âœ… SUCCESS: Removed {deleted_count} leaking images from Training set.")
    print("Your dataset is now scientifically valid.")
else:
    print("âœ… No duplicates found. You are safe.")
```
